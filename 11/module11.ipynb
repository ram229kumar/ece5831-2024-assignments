## Binary Classification - IMDB

```python
import numpy as np
from keras.datasets import imdb
from imdb import Imdb
```

```python
imdb_obj = Imdb()
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=imdb_obj.DIMENSIONS)
```

```python
x_train = imdb_obj.prepare_data(x_train)
x_test = imdb_obj.prepare_data(x_test)
```

```python
model = imdb_obj.build_model()
```

```python
history = imdb_obj.train(x_train, y_train)
```

```python
imdb_obj.plot_loss(history)
```

```python
imdb_obj.plot_accuracy(history)
```

```python
test_loss, test_accuracy = imdb_obj.evaluate(x_test, y_test)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
print("\nPrediction Example:")
imdb_obj.predict(model, x_test[0], y_test[0])
```

## Multiclass Classification - Reuters

```python
from reuters import Reuters
from keras.datasets import reuters
from keras.utils import to_categorical
import numpy as np
```

```python
reuters_obj = Reuters()
```

```python
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=reuters_obj.DIMENSIONS)
```

```python
x_train = reuters_obj.prepare_data(x_train)
x_test = reuters_obj.prepare_data(x_test)
```

```python
model = reuters_obj.build_model()
```

```python
history = reuters_obj.train(x_train, y_train)
```

```python
reuters_obj.plot_loss(history)
```

```python
reuters_obj.plot_accuracy(history)
```

```python
test_loss, test_accuracy = reuters_obj.evaluate(x_test, to_categorical(y_test, num_classes=reuters_obj.NUM_CLASSES))
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_accuracy}")
```

```python
reuters_obj.predict(x_test[0],y_test[0])
```

## Regression - Boston Housing

```python
from keras.datasets import boston_housing
from boston_housing import BostonHousing
```

```python
boston_housing_obj = BostonHousing()
```

```python
(x_train, y_train), (x_test, y_test) = boston_housing.load_data()
```

```python
x_train = boston_housing_obj.prepare_data(x_train)
x_test = boston_housing_obj.prepare_data(x_test)
```

```python
model = boston_housing_obj.build_model()
```

```python
history = boston_housing_obj.train(x_train, y_train)
```

```python
boston_housing_obj.plot_loss(history)
```

```python
test_mse_score,test_mae_score = boston_housing_obj.evaluate(x_test, y_test)
print(f"Test MSE Score: {test_mse_score}")
print(f"Test MAE Score: {test_mae_score}")
```
